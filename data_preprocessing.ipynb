# Adding Leak column

```python
lcols=[]
for i in range(32):
	col= f'Leak_Node-{i+1}'
	lcols.append(col)
lcols
```

```python
def leak_detect(sum):
    if(sum==0):
        return 0
    else:
        return 1

df['Leak']=df[lcols].sum(axis=1).apply(lambda x: leak_detect(x))
```

# Removing unwanted columns

```python
df.drop(['Unnamed: 0', 'Timestamp', 'Scenario'], axis=1, inplace=True)
```

# Normalising

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = float(df[feature_name].max())
        min_value = float(df[feature_name].min())
        df[feature_name].apply(lambda x: float(x))
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
df_normalized = normalize(df)
```
