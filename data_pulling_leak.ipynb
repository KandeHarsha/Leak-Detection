# Final Dataframe

```python
df1 = pd.read_csv(f"{endpoint1}/Pressure_csv_file.csv")
df2 = pd.read_csv(f"{endpoint1}/Flows_csv_file.csv")
df3 = pd.read_csv(f"{endpoint1}/Leak_csv_file.csv")
```

```python
merged_df = df1.merge(df2, on='Unnamed: 0').merge(df3, on='Unnamed: 0')
merged_df
```

```python
lcols=[]
for i in range(32):
	col= f'leak-{i+1}'
	lcols.append(col)
lcols
```

```python
def leak_detect(sum):
    if(sum==0):
        return 0
    else:
        return 1

merged_df['Leak']=merged_df[lcols].sum(axis=1).apply(lambda x: leak_detect(x))
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = float(df[feature_name].max())
        min_value = float(df[feature_name].min())
        df[feature_name].apply(lambda x: float(x))
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
df_normalized = normalize(merged_df)
df_normalized
```
